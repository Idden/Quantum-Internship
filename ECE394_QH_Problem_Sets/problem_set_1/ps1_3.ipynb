```python
import qutip as qt
import numpy as np
```

```python
# problem 3 part b
# constants
hbar = 1
omega = 1
mass = 1
lamb = 1

# creation and annihilation operators
N = 500
adag = qt.create(N)
a = qt.destroy(N)

# x and p ops in creation and annihilation basis
x_op = (a + adag) * np.sqrt(hbar / 2 / mass / omega)
p_op = 1j * (adag - a) * np.sqrt(mass * omega * hbar / 2)

# create hamiltonian
kineticTerm = p_op**2 / 2 / mass
potentialTerm = 0.5 * mass * omega**2 * x_op**2
perturbationTerm = lamb / 24 * x_op**4
H = kineticTerm + potentialTerm + perturbationTerm

# diagonalize
eigenvalues, eigenstates = H.eigenstates()
```

```python
# problem 3 part c
# find convergence
n_list = np.linspace(10, 100, 10)
for n in n_list:
    n = int(n)

    # creation and annihilation operators for n basis size
    adag = qt.create(n)
    a = qt.destroy(n)

    # x and p ops in creation and annihilation basis
    x_op = (a + adag) * np.sqrt(hbar / 2 / mass / omega)
    p_op = 1j * (adag - a) * np.sqrt(mass * omega * hbar / 2)

    # create hamiltonian
    kineticTerm = p_op**2 / 2 / mass
    potentialTerm = 0.5 * mass * omega**2 * x_op**2
    perturbationTerm = lamb / 24 * x_op**4
    H = kineticTerm + potentialTerm + perturbationTerm

    # diagonalize
    eigenvalues, eigenstates = H.eigenstates()

    print(f"Eigenvalues for N={n} hilberty space")
    for i in range(6):
        print(eigenvalues[i])
    print()
```
