```python
import scqubits as scq
import numpy as np
import matplotlib.pyplot as plt
```

```python
tmon = scq.Transmon.create()

transmon1 = scq.Transmon(
    EJ=30.0,
    EC=1.2,
    ng=0.3,
    ncut=31
)

transmon2 = scq.Transmon(
    EJ=30.0,
    EC=1.2,
    ng=1.5,
    ncut=31
)

eval, evecs = tmon.eigensys(evals_count=3)
transmon1.plot_n_wavefunction(esys=None, which=0, mode='real')
transmon2.plot_n_wavefunction(esys=None, which=0, mode='real')
```

```python
tmon.plot_phi_wavefunction(esys=None, which=[0,1,2,3,4], mode='real');
```

```python
ng_list = np.linspace(-2, 2, 220)
transmon1.plot_evals_vs_paramvals('ng', ng_list, evals_count=5, subtract_ground=True)
```

```python
EJvals = np.linspace(0.1, 35, 100)
tmon.plot_dispersion_vs_paramvals('ng', 'EJ', EJvals, ref_param='EC', levels=(0,1,2))
tmon.plot_dispersion_vs_paramvals('ng', 'EJ', EJvals, ref_param='EC', transitions=(((0,1), (0,2))))
```
