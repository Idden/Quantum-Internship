```python
import qutip as qt
import numpy as np
import matplotlib.pyplot as plt
```

```python
wq = 0.642570281124498
wd = 0.642570281124498
A = 0.1

sx = qt.sigmax()
sy = qt.sigmay()
sz = qt.sigmaz()

H0 = wq/2*sz
H1 = sx
eigvals, eigvecs = H0.eigenstates()

def drive(t, args):
    return A*np.cos(wd*t)

H = [H0, [H1, drive]]

psi0 = qt.basis(2,0)
tlist = np.linspace(0, 50, 200)
psi_t = qt.sesolve(H, psi0, tlist, e_ops=[sz])

plt.plot(tlist, psi_t.expect[0])
plt.show()
```
